```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import requests
from bs4 import BeautifulSoup

import re
import time

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# WEB SCRAPING OF IPL DATA BY USING URL THROUGH INSPECTING HTML TAGS
# BY USING REQUESTS LIBRARY
# BY USING BEAUTIFUL SOUP
```

```python
player=[]
matches=[]
innings=[]
Notouts=[]
mruns=[]
highestscore=[]
bestbattingaverage=[]
ballsfaced=[]
strikerate=[]
hundreds=[]
mfifties=[]
msixes=[]
mfours=[]
```

```python
a=["most-runs","most-sixes","highest-scores","best-batting-strike-rate","best-batting-average","most-fifties","most-fours","fastest-fifties","fastest-centuries"]
for i in a:
    
    url="https://www.iplt20.com/stats/all-time/{}".format(i)
    page = requests.get(url)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    for plyr in soup.find_all("div",{"class":"top-players__player-name"}):
        player.append(plyr.text)
    for mach in soup.find_all("td",{"class":"top-players__m top-players__padded"}):
        matches.append(mach.text)
    for inn in soup.find_all("td",{"class":"top-players__inns"}):
        innings.append(inn.text)
    for nouts in soup.find_all("td",{"cla ss":"top-players__no"}):
        Notouts.append(nouts.text)
    for run in soup.find_all("td",{"class":"top-players__r"}):
        mruns.append(run.text)
    for hs in soup.find_all("td",{"class":"top-players__hs"}):
        highestscore.append(hs.text)
    for Avg in soup.find_all("td",{"class":"top-players__a"}):
        bestbattingaverage.append(Avg.text)
    for balfc in soup.find_all("td",{"class":"top-players__b"}):
        ballsfaced.append(balfc.text)
    for sr in soup.find_all("td",{"class":"top-players__sr"}):
        strikerate.append(sr.text)
    for hun in soup.find_all("td",{"class":"top-players__100s"}):
        hundreds.append(hun.text)
    for fiftiy in soup.find_all("td",{"class":"top-players__50s"}):
        mfifties.append(fiftiy.text)
    for sxs in soup.find_all("td",{"class":"top-players__6s"}):
        msixes.append(sxs.text)
    for furs in soup.find_all("td",{"class":"top-players__4s"}):
        mfours.append(furs.text)
    
```

```python
batsmen=[]
a=["most-runs","most-sixes","highest-scores","best-batting-strike-rate","best-batting-average","most-fifties","most-fours"]
for i in a:
    
    url="https://www.iplt20.com/stats/all-time/{}".format(i)
    page = requests.get(url)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    for batsmn in soup.find_all("div",{"class":"top-players__player-name"}):
        batsmen.append(batsmn.text)
```
