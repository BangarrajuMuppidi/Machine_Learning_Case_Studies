## Function for all degree

```python
def polyregress(*degree):
    
    
    for i in degree:
        
        ## using polynomial futures
        
        from sklearn.preprocessing import PolynomialFeatures

        poly_deg_2 = PolynomialFeatures(degree=i)
        
        ## poynomial transformation of X_train

        X_train_pd2 = poly_deg_2.fit_transform(X_train)

        model_pd2 = LinearRegression()
        
        ### Fitting the model
 
        model_pd2.fit(X_train_pd2, Y_train)
       
    
        ### Residual analysis on train data
        
        y_train_pred_pd2=model_pd2.predict(X_train_pd2)
        
        ### resuidal error on train
        
        residual_train_pd2=Y_train-y_train_pred_pd2
        
        ## Checking resuidal error with displot
        sns.displot(residual_train_pd2,kind="kde",color="orange")
        
        plt.show()
        ## checking Homoscadasticity with scatterplot
        
        plt.scatter(Y_train, residual_train_pd2)

        plt.show()
        
        ### polynomial transformation of X_test
        
        X_test_pd2 = poly_deg_2.transform(X_test)
        
        ### predection

        y_test_pred_pd2 = model_pd2.predict(X_test_pd2)
        
        print("Degree======>",i)
        
        ### Evalution of metrics
        
        from sklearn import metrics

        print('Mean Absolute Error: ', metrics.mean_absolute_error(Y_test, y_test_pred_pd2))
        print('Mean Squared Error: ', metrics.mean_squared_error(Y_test, y_test_pred_pd2))
        print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(Y_test, y_test_pred_pd2)))
        print('R2 score:',metrics.r2_score(Y_test, y_test_pred_pd2))
        print("==============================================================================")
        
             
polyregress(1,2,3,4,5,6,7)
```

```python
def polynomial_feature(*degree):
    for i in degree:
        #polynomial transformation in X_train

        from sklearn.preprocessing import PolynomialFeatures

        poly_deg_2 = PolynomialFeatures(degree=i)

        X_train_pd2 = poly_deg_2.fit_transform(X_train)

        #Fit the model
        model_pd2 = LinearRegression()

        model_pd2.fit(X_train_pd2, Y_train)

        # residual analysis
        Y_train_pred = model_pd2.predict(X_train_pd2);
        residual = Y_train - Y_train_pred

        #Prediction
        X_test_pd2 = poly_deg_2.transform(X_test)
        Y_test_pred_pd2 = model_pd2.predict(X_test_pd2)
        residual_test= Y_test - Y_test_pred_pd2
        
        #Model-1 boxplot
        plt.figure(figsize=(25,5))
        plt.subplot(2,4,i)
        sns.boxplot(x=residual_test,color="grey");
        print('\033[1m' + f'R-square value for polynomial degree {i}: ' + '\033[0m ',metrics.r2_score(Y_test, Y_test_pred_pd2))

polynomial_feature(1,2,3,4,5,6,7)
```
